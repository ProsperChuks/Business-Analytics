### Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pickle
```

### Loading the Data

```python
df = pd.read_csv('assessment.csv')
df
```

### Preprocessing

#### Handling Missing Values

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
cate = df.Product_Category.drop_duplicates()
cate.reset_index(drop=True, inplace=True)
cate
```

```python
electric = df.loc[df.Product_Category == cate[0]]
telecom = df.loc[df.Product_Category == cate[1]]
cable = df.loc[df.Product_Category == cate[2]]
pin = df.loc[df.Product_Category == cate[3]]
edu = df.loc[df.Product_Category == cate[4]]
```

#### Electricity

```python
electric = electric[['user_id', 'TotalTransactionAmount', 'Tx_Count', 'DaysSinceLastTrans']]
electric
```

```python
elec_customers = electric.groupby(['user_id']).agg({
    'TotalTransactionAmount': 'sum',
    'Tx_Count': 'count',
    'DaysSinceLastTrans' : 'sum'
})
```

```python
elec_customers.rename(columns={'TotalTransactionAmount': 'MV',
                                 'Tx_Count': 'Freq',
                                'DaysSinceLastTrans': 'Re'}, inplace=True)
elec_customers
```

##### Converting Data to Gaussian Distribution

```python
print(elec_customers.Re.skew())
print(elec_customers.Freq.skew())
print(elec_customers.MV.skew())
```

```python
import warnings
warnings.filterwarnings('ignore')

fig, ax = plt.subplots(2, 2, figsize=(12, 7))

fig = sns.distplot(elec_customers.Re, color='b', ax=ax[0, 0])
fig = sns.distplot(elec_customers.MV, color='g', ax=ax[0, 1])
fig = sns.distplot(elec_customers.Freq, color='y', ax=ax[1, 0])
ax[0, 0].set_title('Recency')
ax[0, 1].set_title('Monetary Value')
ax[1, 0].set_title('Frequency')

plt.tight_layout()
plt.show()
```

```python
elec_customers['MV'] = stats.boxcox(elec_customers['MV'])[0]
elec_customers['Freq'] = stats.boxcox(elec_customers['Freq'])[0]
elec_customers['Re'] = stats.boxcox(elec_customers['Re'])[0]
elec_customers
```

```python
print(elec_customers.Re.skew())
print(elec_customers.Freq.skew())
print(elec_customers.MV.skew())
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 7))

fig = sns.distplot(elec_customers.Re, color='b', ax=ax[0, 0])
fig = sns.distplot(elec_customers.MV, color='g', ax=ax[0, 1])
fig = sns.distplot(elec_customers.Freq, color='y', ax=ax[1, 0])
ax[0, 0].set_title('Recency')
ax[0, 1].set_title('Monetary Value')
ax[1, 0].set_title('Frequency')

plt.tight_layout()
plt.show()
```

```python
elec_customers.describe()
```

##### Normalization

```python
scale = StandardScaler()
scale.fit(elec_customers)
electric_normalized = scale.transform(elec_customers)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 7))

fig = sns.distplot(electric_normalized[0], color='b', ax=ax[0, 0])
fig = sns.distplot(electric_normalized[1], color='g', ax=ax[0, 1])
fig = sns.distplot(electric_normalized[2], color='y', ax=ax[1, 0])
ax[0, 0].set_title('Recency')
ax[0, 1].set_title('Monetary Value')
ax[1, 0].set_title('Frequency')

plt.tight_layout()
plt.show()

print(electric_normalized.mean(axis = 0).round(2))
print(electric_normalized.std(axis = 0).round(2))
```

### Customer Segementation Analysis (Electricity)

```python
model = KMeans(n_clusters=3, random_state=42)
model.fit(electric_normalized)
model.labels_.shape
```

```python
elec_customers["Cluster"] = model.labels_
elec_customers.groupby('Cluster').agg({
    'Re':'mean',
    'Freq':'mean',
    'MV':['mean', 'count']}).round(2)
elec_customers
```

```python
df_normalized = pd.DataFrame(electric_normalized, columns=['Re', 'Freq', 'MV'])
df_normalized['ID'] = elec_customers.index
df_normalized['Cluster'] = model.labels_
```

```python
df_nor_melt = pd.melt(df_normalized.reset_index(),
                      id_vars=['ID', 'Cluster'],
                      value_vars=['Re','Freq','MV'],
                      var_name='Attribute',
                      value_name='Value')
df_nor_melt
```

```python
plt.figure(figsize=(15, 8))
sns.lineplot('Attribute', 'Value', hue='Cluster', data=df_nor_melt)
plt.show()
```
